```python
from bs4 import BeautifulSoup as bs
import requests
url = "https://www.imdb.com/search/title/?title_type=feature&release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start=51&ref_=adv_nxt"
response = requests.get(url)
response = bs(response.content, "html.parser")
print(type(response))
```

```python
print(response.p)
```

```python
conteneur_main = response.find(id="main")
films = conteneur_main.find_all(class_="lister-item mode-advanced")
```

```python
"""**Exercice 3:**
* Extraire l'année de sortie
* Extraire la note IMDB (à convertir en nombre à virgule)
* Extraire la note Metacritic (à convertir en entier)
* Extraire le nombre de votes (utiliser paramètre attrs)
"""
year = films[0].find(class_="lister-item-year").get_text()
print(year)

imdb = films[0].find(class_="ratings-imdb-rating").find("strong").get_text()
print(imdb)

metacritic = films[0].find(class_="ratings-metascore").find("span").get_text()
print(metacritic)

votes = films[0].find(class_="sort-num_votes-visible").find_all("span")[1].get_text()
print(votes)

```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

import time
```

```python
"""
Utiliser attrs

soup.find(name="nv") 
soup.find("span", attrs = {"name": "nv"} 
soup.find("span", attrs={"class":"metascore"}
"""
```

```python
code = 200
start = 1
df = pd.DataFrame(columns=['Titre', 'Annee', 'Note', 'Metacritic', 'Votes'])
i=0
while start < 12123:
    url = "https://www.imdb.com/search/title/?title_type=feature&release_date=2017-01-01,2017-12-31&sort=num_votes,desc&start="+ str(start) +"&ref_=adv_nxt"
    response = requests.get(url)
    code = response.status_code
    if code != 200 :
        break
    if start > 12123 :
        break
    page_content = bs(response.content, "html.parser")

    conteneur_main = page_content.find(id="main")
    films = conteneur_main.find_all(class_="lister-item mode-advanced")

    for film in films :


        titre = film.find(class_="lister-item-header").find("a").get_text()
        year = film.find(class_="lister-item-year").get_text()

        votes = film.find(class_="sort-num_votes-visible")
        if votes is not None :
            votes = votes.find_all("span")
            if len(votes) >0:
                votes = votes[1]
                votes = votes.get_text()
            else:
                votes = None

        imdb = film.find(class_="ratings-imdb-rating")
        if imdb is not None :
            imdb = imdb.find("strong").get_text()
            
        metacritic = film.find(class_="ratings-metascore")
        if metacritic is not None :
            metacritic= metacritic.find("span").get_text()
        

        df.loc[i] = [titre] + [year] + [imdb] + [metacritic] + [votes]
        i+=1
        start += 1

    time.sleep(0.5)
    
    print(str(code) + " " + str(start))
print(df)
```
